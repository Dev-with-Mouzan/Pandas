```python
import pandas as pd 
```

```python
df=pd.read_csv("Titanic-Dataset.csv") 
```

```python
# rename the columns names
update_df=df.rename(columns={"PassengerId":"Passenger_ID","Survived":"Survived_or_Not","Pclass":"Passenger_Class",})
# syntax rename(columns={"old_name1":"new_name1","old_name2":"new_name2",...})
```

```python
update_df.columns
```

```python
# Type_Cast
update_df["Survived_or_Not"]=update_df["Survived_or_Not"].astype("bool")
```

```python
update_df["Survived_or_Not"].dtype
```

```python
# Apply method 
# let i want to multiply the age column with 2 so i have to use loop but in pandas we have apply method to do this type of work

update_df["Age"].apply(lambda X: X*2)
 
# It is more faster then loop 

```
